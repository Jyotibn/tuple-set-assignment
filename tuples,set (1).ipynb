```python
def myreduce(add,l):
    a=l[0]
    for i in range(1,len(l)):            
            b=l[i]
            a=add(a,b)
         
    return a


def add(a,b):
    return a+b
        
        
        
    
    
```

```python
myreduce(add,[1,2,3,4,5])
```

```python
def myfilter(even,l):
    l1=[]
    for i in l:
        if even(i)!=None:
            
            l1.append(even(i))
    
    return l1
    
    
def even(a):
    if a%2==0:
        return a
```

```python
myfilter(even,[1,2,3,4,5,6])
```
